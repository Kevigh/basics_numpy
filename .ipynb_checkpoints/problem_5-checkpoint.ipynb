```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#find all the voltages of a circuit

A = np.array([[3, 2, 3, 10],[2, -2, 5, 8],[3, 3, 4, 9],[3, 4, -3, -7]])
b = np.array([4, 1, 3, 2])

V = np.linalg.solve(A, b)

for i in range(4):
    print('V{}:'.format(i + 1), V[i])
```

An electric field is given by $\vec{E}(z,t) = E_{0}\cos(z - t)\hat{x} + 2E_{0}\cos{(z-t+\frac{\pi}{2}})\hat{y}$.

1. Find the magnetic field for $z \in [0,4\pi]$ and $t \in [0,10]$.
2. Compute the Poynting vector $\vec{S} =\vec{E}\times\vec{B}$.

```python
#limits
z = np.linspace(0, 4*np.pi, 1000)
t = np.linspace(0, 10, 1000)

#mesh grid
zv, tv = np.meshgrid(z,t)

#electric field
E_x = np.cos(zv - tv)
E_y = 2*np.cos(zv - tv + np.pi/2)
E_z = 0*tv
```

```python
#plot Ex as a function of t at z = 0

plt.plot(t, E_x[0], 'k')
plt.title('$E_{x}$ as a function of t at z = 0', fontsize = 20)
plt.xlabel('$t$', fontsize = 20)
plt.ylabel(r'$\frac{E(t)}{E_{0}}$', fontsize = 20)
plt.show()

#plot Ex as a function of z at t = 0

plt.plot(t, E_x[:,0], 'k')
plt.title('$E_{x}$ as a function of z at t = 0', fontsize = 20)
plt.xlabel('$z$', fontsize = 20)
plt.ylabel(r'$\frac{E(z)}{E_{0}}$', fontsize = 20)
plt.show()
```

```python
#build the electric field

E_array = np.array([E_x, E_y, E_z])

E = np.swapaxes(E_array, 0, -1)

#find the magnetic field

B = np.cross(np.array([0, 0, 1]), E)

B_array = np.swapaxes(B, 0, -1)

# components of magnetic field

B_x, B_y, B_z = B_array
```

```python
#plot Ex and Bx as a function of t at z = 0

plt.plot(t, B_x[0], 'k')
plt.plot(t, E_y[0], 'r')
plt.show()
```

```python
#poynting vector

S = np.cross(E, B)

S_array = np.swapaxes(S, 0, -1)

Sx, Sy, Sz = S_array
```
