```python
import numpy as np
import matplotlib.pyplot as plt
```

The power emitted by a blackbody is $P = A\sigma\epsilon T^{4}$. After measuring the temperature of a star you find that

$$T(t) = T_{0}\frac{1}{1 + e^{-kt}}.$$

Plot the total energy emitted by the star as a function of time using the fact that $E(t) = \int_{t'=0}^{t'=t} = P(t')dt'.$

We don't know what $A$, $\sigma$, and $\epsilon$ necessarily are.

We can't deal with $P$, so we can define a quantity

$$\frac{P}{A \sigma \epsilon T_{0}^{4}} = \left( \frac{1}{1 + e^{-kt}} \right)^{4}.$$

Then, we can plot this function because is dimensionless

$$D = \left( \frac{1}{1 + e^{-kt}} \right)^{4}.$$

```python
kt = np.linspace(0, 5, 100)
D = (1 / (1 + np.exp(-kt))) ** 4
```

```python
plt.plot(kt, D, 'r', label = 'Power emitted')
plt.legend()
plt.show()
```

```python
E = np.cumsum(D) * (kt[1] - kt[0])

plt.plot(kt, E, 'b', label = 'Energy')
plt.legend()
plt.xlabel('$kt$', fontsize = 15)
plt.ylabel(r'$\left( \frac{k}{A \sigma \epsilon T_{0}^{4} } \right) E(kt)$', fontsize = 15)
plt.show()
```
