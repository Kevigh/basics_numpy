**Integrals and derivatives**

```python
x = np.linspace(1, 10, 100)
y = 1/x**2 * np.sin(x)

plt.plot(x, y, 'r-')
plt.show()
```

```python
#derivative
dydx = np.gradient(y, x)

#integral
Sydx = np.cumsum(y) * (x[1] - x[0])

plt.plot(x, y, 'r-', label = 'Function')
plt.plot(x, dydx, 'b-', label = 'Derivative')
plt.plot(x, Sydx, 'c-', label = 'Integral')
plt.legend()
plt.show()
```
