```python
import numpy as np
import matplotlib.pyplot as plt
```

Let $f(x,y) = e^{-(x^2 + y^2)}\sin(x)$ for $-2\leq x \leq 2$ and $-2\leq y \leq 2$.

```python
x = np.linspace(-2, 2, 1000)
y = np.linspace(-2, 2, 1000)

xv, yv = np.meshgrid(x, y)

zv = np.exp(-(xv**2 + yv**2)) * np.sin(xv)

plt.contourf(xv, yv, zv, levels = 100)
plt.colorbar()
plt.show()
```

```python
#volume of abs(f(x,y))

dx = np.diff(x)[0] #this is equal to x[1] - x[0]
dy = np.diff(y)[0]

V = sum(np.abs(zv.ravel())) * dx * dy

print('Volume:', V)
```

```python
#volume in the region where x**2 + y**2 > 0.5**2

lim = (xv**2 + yv**2) > 0.5**2 #region limit

V = sum(np.abs(zv[lim])) * dx * dy

print('Volume:', V)
```
