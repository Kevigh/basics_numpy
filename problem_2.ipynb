```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sum from 0 to 10000 except for those can be divided by 4 or 7

x = np.arange(0, 10001, 1)

S = sum(x[0:][(x % 4 != 0) * (x % 7 != 0)])

print('Sum:', S)
```

```python
#plot polar, find x and y points

theta = np.linspace(0, 2*np.pi, 1000)

r = 1 + 3/4 * np.sin(3 * theta)

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.plot(x, y, 'r')
plt.show()
```

```python
#compute the area

A = 1/2 * sum(r**2) * (theta[1] - theta[0])

print('Area:', A)
```

```python
#compute the arclength

drd0 = np.gradient(r, theta)

sqrt = np.sqrt(r**2 + (drd0)**2)

L = sum(sqrt) * (theta[1] - theta[0])

print('Arclength:', L)
```
