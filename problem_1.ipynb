```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000
x = np.linspace(0, 10, N)

def f(x):
    return np.exp(-x/10) * np.sin(x)

plt.plot(x, f(x), 'r')
plt.show()
```

```python
#mean
mean = np.mean(f(x[(x >= 4)*(x <= 7)]))
print('Mean:', mean)

#standard deviation
std = np.std(f(x[(x >= 4)*(x <= 7)]))
print('Standard deviation:', std)
```

```python
#percentile

y_m = np.percentile(f(x[(x >= 4)*(x <= 7)]), 80)
print('y_m:', y_m)
```

```python
#make the derivative

dydx = np.gradient(f(x), x)

plt.plot(x, f(x), 'r', label = 'Function')
plt.plot(x, dydx, 'b', label = 'Derivative')
plt.show()
```

```python
#find when f'(x) = 0

e_1 = dydx[1:]
e_2 = dydx[:-1]

p1, p2, p3 = x[1:][e_1 * e_2 < 0] #only when the derivative change sign (+)*(-) = (-)

plt.plot(x, f(x), 'r', label = 'Function')
plt.plot(x, dydx, 'b', label = 'Derivative')
plt.axvline(p1)
plt.axvline(p2)
plt.axvline(p3)
plt.legend()
plt.show()
```
